## Salvando no Delta Lake

```python
%sql
SELECT current_catalog(), current_schema();
```

```python
%sql
-- criando um schema bronze para salvar os dados em Delta
CREATE SCHEMA IF NOT EXISTS workspace.bronze;
USE SCHEMA bronze;
```

```python
df_taxa_brut_nat.write \
.format("delta") \
.mode("overwrite") \
.option("overwriteSchema", "True") \
.saveAsTable("workspace.bronze.taxa_bruta_natalidade")

df_esp_vida_nasc.write \
.format("delta") \
.mode("overwrite") \
.option("overwriteSchema", "True") \
.saveAsTable("workspace.bronze.esperanca_vida_nascer")

print("Tabelas criadas corretamente.")
```
